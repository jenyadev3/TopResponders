```python
df = df.drop(df[df["work_experience"] == "2867959.0"].index)
df = df.drop(df[df["residence_duration"] == "28101997.0"].index)
```

*Определение границ аномалий*

```python
df["work_experience"].info()
```

Определение процентелей

```python
Q1 = df["work_experience"].quantile(0.25)
Q3 = df["work_experience"].quantile(0.75)
```

Межквартальный размах

```python
IQR = Q3 - Q1
```

Определение границ аномалий

```python
lower_level = Q1 - 1.5 * IQR
print(lower_level)
```

```python
upper_level = Q3 + 1.5 * IQR
print(upper_level)
```

Фильтрация аномалий

```python
anomalies = df[(df['work_experience'] < lower_level) | (df['work_experience'] > upper_level)]
print(anomalies["work_experience"])
```

```python
df[["work_experience","age"]].dropna().sort_values(by="work_experience", ascending=False).head(50)

```

Считаем возраст начала работы

```python
df["work_ages"] = df["work_experience"]/12
df["start_age"] = df["age"] - (df["work_ages"])
df["start_age"]
```

Находим тех кто начал раньше 18

```python
df["age_anom"] = df["start_age"] < 18
df["age_anom"]
```

Отбираем только аномальные записи, сортируем по возрасту начала работы от самого большого

```python
age_anom1 = df[df["age_anom"]][["age","work_ages","start_age"]]
age_anom1 = age_anom1.sort_values(by="start_age")
age_anom1.head(50)
```

Фильтруем (удаляем тех кто начал работать меньше чем с 18)

```python
df_clean = df[df["start_age"] >= 18].copy()
print(df_clean[["age","work_ages","start_age","is_employed","is_pensioner","gender"]].head(50))
print("Количество записей при очистке:",len(df_clean))
```
